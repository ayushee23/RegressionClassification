```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn import preprocessing
from pandas.tools.plotting import scatter_matrix
from pandas.stats.api import ols
matplotlib.style.use('ggplot')
import statsmodels.formula.api as sm
import statsmodels.stats.api as sms
```

```python
bp_data = pd.read_csv('Ex03_SystolicBP_Regreesion.csv')
```

```python
bp_data.describe()
```

```python
correlations = bp_data.corr()
fig = plt.figure()
names = bp_data.columns
ax = fig.add_subplot(1,1,1)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(names),1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
correlations
```

```python
scatter_matrix(bp_data, diagonal='kde', figsize=(10,8))
plt.show()
```

```python
splitend = int(0.8 * len(bp_data))
bp_data_shuffled = bp_data.sample(frac=1)
train, test = bp_data_shuffled[:splitend], bp_data_shuffled[splitend:]
x_train = train.ix[:, :-1]
y_train = train.ix[:, -1]
x_test = test.ix[:, :-1]
y_test = test.ix[:, -1]
```

## Model with all features

```python
model1 = sm.ols(formula='SystolicBP~Age+Treated+Gender+FamilySize+Diet+Weight+Height+ScaledBMI', data=train).fit()
```

```python
model1.summary()
```

```python
predictions1 = model1.predict(x_test)
```

```python
np.mean((y_test - predictions1)**2)**0.5
```

```python
plt.scatter(y_test, predictions1)
plt.plot(y_test, predictions1)
plt.show()
```

```python
residuals = y_test - predictions1
```

```python
plt.hist(residuals, bins=5)
plt.show()
```

```python
residuals.plot(kind='box')
plt.show()
```

```python
name = ['F statistic', 'p-value']
test = sms.het_goldfeldquandt(model1.resid, model1.model.exog)
list(zip(name, test))
```
